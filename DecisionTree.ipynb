# Decision Tree

The goal of this notebook is to show how decision trees improve (or overfit) as we increase their depth. For this purpose we use a data set like those in *PolynomialRegression.ipynb*.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from sklearn import tree
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.sans-serif": ["Computer Modern Sans Serif"],
    "font.size": 14,
    "figure.dpi": 300})
```

## Generating the data set

```python
X_func = np.linspace(0, 1, 1001, endpoint=True)
y_func = np.sin(2 * np.pi * X_func)
```

```python
## Uncomment to generate new random observations of training data
#y_train = (np.sin(2 * np.pi * X_train) + np.random.normal(scale=0.25, size=500).reshape(1, -1)).T
#np.savetxt('InputData/DTs-sine_dataset.txt', y_train)

X_train = np.linspace(0, 1, 500, endpoint=True)
y_train = np.loadtxt('InputData/DTs-sine_dataset.txt')
```

```python
plt.scatter(x=X_train, y=y_train, s=20, marker="o")
plt.plot(X_func, y_func, linewidth=3, linestyle='dashed', color='black')
plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')
plt.ylim((-1.25, 1.25))
plt.legend([r"Training data", r'$\sin(2\pi x)$'])
plt.show()
```

## Decision Tree fits

### Depth = 1

```python
X_train = X_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
```

```python
clf = tree.DecisionTreeRegressor(max_depth=1)
clf = clf.fit(X_train, y_train)
pred1 = clf.predict(X_func.reshape(-1,1))
```

```python
#tree.plot_tree(clf)
```

```python
plt.scatter(x=X_train, y=y_train, s=20, marker="o")
plt.plot(X_func, y_func, linewidth=3, linestyle='dashed', color='black')
plt.plot(X_func, pred1, linewidth=3, color='C1')
plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')
plt.ylim((-1.25, 1.25))
plt.legend([r"Training data", r'$\sin(2\pi x)$', r'DT ${\rm depth}=1$'])
plt.savefig('Plots/DT-depth1.png', bbox_inches='tight', pad_inches=0.01)
plt.show()
```

### Depth = 2

```python
clf = tree.DecisionTreeRegressor(max_depth=2)
clf = clf.fit(X_train, y_train)
pred2 = clf.predict(X_func.reshape(-1,1))
```

```python
#tree.plot_tree(clf)
```

```python
plt.scatter(x=X_train, y=y_train, s=20, marker="o")
plt.plot(X_func, y_func, linewidth=3, linestyle='dashed', color='black')
plt.plot(X_func, pred2, linewidth=3, color='C1')
plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')
plt.ylim((-1.25, 1.25))
plt.legend([r"Training data", r'$\sin(2\pi x)$', r'DT ${\rm depth}=2$'])
plt.savefig('Plots/DT-depth2.png', bbox_inches='tight', pad_inches=0.01)
plt.show()
```

### Depth = 3

```python
clf = tree.DecisionTreeRegressor(max_depth=3)
clf = clf.fit(X_train, y_train)
pred3 = clf.predict(X_func.reshape(-1,1))
```

# Gradient Boosting

We now solve the same problem we solved in *DecisionTree.ipynb* with gradient boosting. More specifically, we show how the fit to a sine function improves by increasing the number of trees in the ensemble.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from sklearn import tree
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.dummy import DummyRegressor
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.sans-serif": ["Computer Modern Sans Serif"],
    "font.size": 14,
    "figure.dpi": 300})
```

## The data set

```python
X_func = np.linspace(0, 1, 1001, endpoint=True)
y_func = np.sin(2 * np.pi * X_func)
```

```python
##The code to generate the data set is already shown in DecisionTree.ipynb.
X_train = np.linspace(0, 1, 500, endpoint=True)
y_train = np.loadtxt('InputData/DTs-sine_dataset.txt')
```

```python
plt.scatter(x=X_train, y=y_train, s=20, marker="o")
plt.plot(X_func, y_func, linewidth=2, linestyle='dashed', color='black')
plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')
plt.ylim((-1.25, 1.25))
plt.legend([r"Training data", r'$\sin(2\pi x)$'])
plt.show()
```

```python
X_train = X_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
```

```python
dummy_regr = DummyRegressor(strategy="mean") # Created to make an initial prediction
dummy_regr.fit(X_train, y_train)
```

## Gradient Boosting fits

### Number of trees = 1

```python
reg1 = GradientBoostingRegressor(learning_rate=0.1, n_estimators=1, max_depth=2, criterion='squared_error', validation_fraction=0.1, init=dummy_regr)
reg1.fit(X_train, y_train)
pred1 = reg1.predict(X_func.reshape(-1,1))
```

```python
plt.scatter(x=X_train, y=y_train, s=20, marker="o")
plt.plot(X_func, y_func, linewidth=3, linestyle='dashed', color='black')
plt.plot(X_func, pred1, linewidth=3, color='C1', alpha=1.0)
plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')
plt.ylim((-1.25, 1.25))
plt.legend([r"Training data", r'$\sin(2\pi x)$', r'$F_1(\mathbf{x})$'])
plt.savefig('Plots/GB-fit1.png', bbox_inches='tight', pad_inches=0.01)
plt.show()
```

### Number of trees = 10

```python
reg2 = GradientBoostingRegressor(learning_rate=0.1, n_estimators=10, max_depth=2, criterion='squared_error', validation_fraction=0.1, init=dummy_regr)
reg2.fit(X_train, y_train)
pred2 = reg2.predict(X_func.reshape(-1,1))
```

```python
plt.scatter(x=X_train, y=y_train, s=20, marker="o")
plt.plot(X_func, y_func, linewidth=3, linestyle='dashed', color='black')
plt.plot(X_func, pred2, linewidth=3, color='C1', alpha=1.0)
plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')
plt.ylim((-1.25, 1.25))
plt.legend([r"Training data", r'$\sin(2\pi x)$', r'$F_{10}(\mathbf{x})$'])
plt.savefig('Plots/GB-fit10.png', bbox_inches='tight', pad_inches=0.01)
plt.show()
```

### Number of trees = 25

```python
reg3 = GradientBoostingRegressor(learning_rate=0.1, n_estimators=25, max_depth=2, criterion='squared_error', validation_fraction=0.1, init=dummy_regr)
reg3.fit(X_train, y_train)
pred3 = reg3.predict(X_func.reshape(-1,1))
```

```python
plt.scatter(x=X_train, y=y_train, s=20, marker="o")
plt.plot(X_func, y_func, linewidth=3, linestyle='dashed', color='black')
plt.plot(X_func, pred3, linewidth=3, color='C1', alpha=1.0)
plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')
plt.ylim((-1.25, 1.25))
plt.legend([r"Training data", r'$\sin(2\pi x)$', r'$F_{25}(\mathbf{x})$'])
plt.savefig('Plots/GB-fit25.png', bbox_inches='tight', pad_inches=0.01)
plt.show()
```

### Number of trees = 50

```python
reg4 = GradientBoostingRegressor(learning_rate=0.1, n_estimators=50, max_depth=2, criterion='squared_error', validation_fraction=0.1, init=dummy_regr)
reg4.fit(X_train, y_train)
pred4 = reg4.predict(X_func.reshape(-1,1))
```

```python
plt.scatter(x=X_train, y=y_train, s=20, marker="o")
plt.plot(X_func, y_func, linewidth=3, linestyle='dashed', color='black')
plt.plot(X_func, pred4, linewidth=3, color='C1', alpha=1.0)
plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')
plt.ylim((-1.25, 1.25))
plt.legend([r"Training data", r'$\sin(2\pi x)$', r'$F_{50}(\mathbf{x})$'])
plt.savefig('Plots/GB-fit50.png', bbox_inches='tight', pad_inches=0.01)
plt.show()
```
